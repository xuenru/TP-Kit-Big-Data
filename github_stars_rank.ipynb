```python
from bs4 import BeautifulSoup
import requests
from multiprocessing import Pool
import pandas as pd
import random
```

```python
def get_page_soup(url):
    return BeautifulSoup(requests.get(url).content, 'html.parser')


def get_all_user(url='https://gist.github.com/paulmillr/2657075'):
    soup = get_page_soup(url)

    return[tr.find('a').text for tr in soup.select('article table tbody tr')]


def get_user_mean_stars(username):
    print(f'Getting repos info of user: {username}')
    url = f'https://api.github.com/users/{username}/repos'
    headers = {'Authorization': 'token 92da1a59b63180cdbf6139724657ab35f21dbbab'}
    resp = requests.get(url, headers=headers)
    if resp.status_code != 200:
        print('error:', resp.content)
    repos = resp.json()
    nb_repos = len(repos)
    if nb_repos == 0:
        return 0
    return {'username': username, 'mean_stars': sum(repo['stargazers_count'] for repo in repos) / nb_repos}
#     return {'username': username, 'mean_stars':random.randint(1,100)} #mock


def get_user_sorted_by_stars(all_users):
    p = Pool(processes=2)
    return p.map(get_user_mean_stars, all_users)
```

```python
all_users = get_all_user()
```

```python
user_list = get_user_sorted_by_stars(all_users)
```

```python
user_list = [e for e in user_list if e != 0]
```

```python
users_sorted_by_stars = pd.DataFrame(user_list).sort_values('mean_stars', ascending=False)
users_sorted_by_stars
```
